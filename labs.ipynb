```python
#file2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
#file3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
# Your code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
pd.options.display.max_rows = 100

file1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
file2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
file3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
#cleaning file 1 

file1_clean = file1.rename(columns={'ST': 'State'})
file1_clean['Customer Lifetime Value']= file1_clean['Customer Lifetime Value'].str.replace("%", " ")
file1_clean.dropna(subset=['Customer Lifetime Value'], inplace=True)
file1_clean.columns = file1_clean.columns.str.lower()
file1_clean.columns = file1_clean.columns.str.replace(' ', '_')
file1_clean['customer_lifetime_value'] = file1_clean['customer_lifetime_value'].str.replace('%', '')
file1_clean['customer_lifetime_value'] = pd.to_numeric(file1_clean['customer_lifetime_value'])
file1_clean['number_of_open_complaints'] = file1_clean['number_of_open_complaints'].str.split('/')
file1_clean['number_of_open_complaints'] = file1_clean['number_of_open_complaints'].str[1].astype(int)
file1_clean
```

```python
#cleaning file 2 

file2_clean = file2.rename(columns={'ST': 'State'})
file2_clean['Customer Lifetime Value']= file2_clean['Customer Lifetime Value'].str.replace("%", " ")
file2_clean.dropna(subset=['Customer Lifetime Value'], inplace=True)
file2_clean.columns = file2_clean.columns.str.lower()
file2_clean.columns = file2_clean.columns.str.replace(' ', '_')
file2_clean['customer_lifetime_value'] = file2_clean['customer_lifetime_value'].str.replace('%', '')
file2_clean['customer_lifetime_value'] = pd.to_numeric(file1_clean['customer_lifetime_value'])
file2_clean['number_of_open_complaints'] = file2_clean['number_of_open_complaints'].str.split('/')
file2_clean['number_of_open_complaints'] = file2_clean['number_of_open_complaints'].str[1].astype(int)
file2_clean
```

```python
#cleaning file 3 
file3_clean = file3
file3_clean.columns = file3_clean.columns.str.lower()
file3_clean.columns = file3_clean.columns.str.replace(' ', '_')
file3_clean
```

```python
customer_data = pd.concat([file1_clean,file2_clean,file3_clean], axis = 0)
customer_data
```
